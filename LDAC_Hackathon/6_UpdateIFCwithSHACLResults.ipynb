```python
import ifcopenshell
import ifcopenshell.api
from rdflib import Graph, Namespace
import re

# Load the IFC file
ifc_file = ifcopenshell.open("data/wallsTest.ifc")

# Load the SHACL validation report Turtle file
g = Graph()
g.parse("output/validation_report.ttl", format="turtle")

# Define namespaces
SH = Namespace("http://www.w3.org/ns/shacl#")

# Create a dictionary to map wall guids to their response data
wall_data = {}

# Extract result messages from the SHACL validation report
for result in g.subjects(predicate=SH.resultMessage, object=None):
    message = str(g.value(result, SH.resultMessage))
    
    # Extract GUID and validation result from the message
    match = re.search(r'wall with GUID: (\w+). Result: (True|False)', message)
    if match:
        guid = match.group(1)
        print(match.group(2))
        is_sufficient = match.group(2) == "True"
        wall_data[guid] = {"IsSufficient": is_sufficient}
        print("valRepPass: " + str(is_sufficient))

# Iterate over the walls in the IFC file
for wall in ifc_file.by_type("IfcWall"):
    # Get the wall guid
    wall_guid = wall.GlobalId
    #print(wall_guid)
    # Check if the wall guid exists in the response data
    if wall_guid in wall_data:
        #print("other guid: " + str(wall_guid))
        response = wall_data[wall_guid]

        # Add the "IsSufficient" property to the wall
        pset = ifcopenshell.api.run("pset.add_pset", ifc_file, product=wall, name="UValueCheck")
        ifcopenshell.api.run("pset.edit_pset", ifc_file, pset=pset, properties={"IsSufficient": response["IsSufficient"]})

        # Change the color of the wall if it fails
#        if not response["IsSufficient"]:
#            representation = ifcopenshell.api.run("style.add_style", ifc_file, style=None)
#            ifcopenshell.api.run("style.set_surface_colour", ifc_file, style=representation, colour=(1, 0, 0))
#            ifcopenshell.api.run("style.assign_representation_styles", ifc_file, representation=representation, products=[wall])

# Save the updated IFC file
ifc_file.write("output/wallsTest_updated_SHACL.ifc")
```
